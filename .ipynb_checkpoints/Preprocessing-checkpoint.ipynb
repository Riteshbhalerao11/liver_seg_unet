```python
from glob import glob 
import os

import torch
from monai.transforms import (
    Compose,
    LoadImaged,
    ToTensord,
    AddChanneld,
    Spacingd,
    ScaleIntensityRanged,
    CropForegroundd,
    Resized,
)

from monai.data import Dataset, DataLoader
from monai.utils import first
import matplotlib.pyplot as plt
```

```python
data_dir = "D:\Kaggle\Datasets\Liver Data"
```

```python
train_images = sorted(glob(os.path.join(data_dir,"train\data","*.nii")))
train_labels = sorted(glob(os.path.join(data_dir,"train\labels","*.nii")))

test_images = sorted(glob(os.path.join(data_dir,"test\data","*.nii")))
test_labels = sorted(glob(os.path.join(data_dir,"test\labels","*.nii")))
```

```python
train_files = [{'image':image_name, 'label' : label_name } for image_name,label_name in zip(train_images,train_labels)]
test_files = [{'image':image_name, 'label' : label_name } for image_name,label_name in zip(test_images,test_labels)]
```

```python
train_transforms = Compose (
    [
        LoadImaged(keys=['image','label']),
        AddChanneld(keys=['image','label']),
        Spacingd(keys=["image","label"], pixdim=(1.5,1.5,2)),
        ScaleIntensityRanged(keys=["image"],a_min=-200,a_max=200,b_min=0.0,b_max=1.0,clip=True),
        CropForegroundd(keys=["image","label"],source_key='image'),
        Resized(keys=['image','label'],spatial_size=[128,128,128]),
        ToTensord(keys=['image','label']),
    ]
)

test_transforms = Compose (
    [
        LoadImaged(keys=['image','label']),
        AddChanneld(keys=['image','label']),
        ToTensord(keys=['image','label']),
    ]
)
```

```python
train_ds = Dataset(data = train_files, transform=train_transforms)
train_loader = DataLoader(train_ds, batch_size=1)

test_ds = Dataset(data = test_files, transform=test_transforms)
test_loader = DataLoader(test_ds, batch_size=1)
```

```python
test_patient  = first(train_loader)
```

```python
plt.figure("test",(12,6))

plt.subplot(1,2,1)
plt.title("Slice of a patient")
plt.imshow(test_patient["image"][0,0,:,:,-20],cmap="gray")

plt.subplot(1,2,2)
plt.title("Label of a patient")
plt.imshow(test_patient["label"][0,0,:,:,-20])

plt.show()
```
